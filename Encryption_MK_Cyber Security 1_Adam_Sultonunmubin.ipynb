```python
# PROGRAM ENKRIPSI & DESKRIPSI
# Adam Sultonunmubin_1462100157

# membuat caesar enkripsi
def caesar_encryption():

  # inisialisasi alfabet, angka, dan simbol
    alpha_upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZABCDEFGHIJKLMNOPQRSTUVWXYZ"
    alpha_lower = "abcdefghijklmnopqrstuvwxyzabcdefghijklmnopqrstuvwxyz"
    symbolic = "~!@#$%^&*()-=+_`}{[]|':;><,./?"
    integer = "1234567890"

    shift = int(input("Masukkan panjang value (0-25) : "))
    text = input("Masukkan plain text: ")
    encrypted_text = ""

  #loopong : mengiterasi setiap karakter di  variabel 'e' dalam 'text'
    for e in text:
      #looping : jika text yang diinputkan huruf kapital
        if e.isupper():
          #maka, e_index : mencari karakter kapital dalam string 'apha_upper'
            e_index = alpha_upper.find(e)
            #lalu new_index menghitung index baru setelah menggeser sebanyak shift dgn modulus/panjang /> 26
            new_index = (e_index + shift) % 26
            #lalu new_char mengambil karakter baru dari 'apha_upper' sesuai index yg dihitung
            new_char = alpha_upper[new_index]
            #lalu new_char menambahkan karakter yang baru tadi kedalam 'encrypted_text
            encrypted_text += new_char

      #looping : jika text yang diinputkan huruf kecil (proses sama dengan huruf kapital)
        elif e.islower():
            e_index = alpha_lower.find(e)
            new_index = (e_index + shift) % 26
            new_char = alpha_lower[new_index]
            encrypted_text += new_char
      #mengecek kondisi karakter var e adalah (spasi)
        elif e == " ":
          #jika ya, maka spasi akan ditambahkan ke 'encrypted_text' tanpa modifikasi
            encrypted_text += " "
      #mengecek kondisi karakter var e adalah (angka numerik)
        elif e.isnumeric():
          #jika ya, maka angka numerik akan ditambahkan ke 'encrypted_text' tanpa modifikasi
            encrypted_text += e
      #proses enkripsi untuk karakter khusus (non-alfanumerik) atau symbolic
        else:
          #jadi, 'e_index' mencari mencari string 'simbolyc' didalam var e
            e_index = symbolic.find(e)
          #lalu, mengambil karakter baru dari 'symbolic' sesuai dengan indeks yg dihitung di var e
            new_char = symbolic[e_index]
          #lalu 'new_char' menambahkannya kedalam 'encrypted_text'
            encrypted_text += new_char

    #setelah itu menampilkan menggunakan print
    print("Berikut adalah hasil Encryption :", encrypted_text)

# membuat caesar deskripsi
def caesar_decryption():
    alpha_upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZABCDEFGHIJKLMNOPQRSTUVWXYZ"
    alpha_lower = "abcdefghijklmnopqrstuvwxyzabcdefghijklmnopqrstuvwxyz"
    symbolic = "~!@#$%^&*()-=+_`}{[]|':;><,./?"
    integer = "1234567890"

    text = input("Masukkan plain text: ")

  #looping : membuat iterasi melalui nilai shift dengan range 0-25
  #          dimana tiap iterasi akan mencoba 1 1 nilai dari shift
    for shift in range(26):
      # menginisiasi ulang untuk menyimpan hasil deskripsi yang diinputkan user dalam text
        decrypted_text = ""
        # membuat looping yang sama dengan enkripsi, bedanya adalah tiap karakter pada text yg
        # diinputkan akan diinisasi ulang sampai sebanyak 25 iterasi
        for d in text:
            if d.isupper():
                d_index = alpha_upper.find(d)
                new_index = (d_index - shift) % 26
                new_char = alpha_upper[new_index]
                decrypted_text += new_char
            elif d.islower():
                d_index = alpha_lower.find(d)
                new_index = (d_index - shift) % 26
                new_char = alpha_lower[new_index]
                decrypted_text += new_char
            elif d == " ":
                decrypted_text += " "
            elif d.isnumeric():
                decrypted_text += d
            else:
                d_index = symbolic.find(d)
                new_char = symbolic[d_index]
                decrypted_text += new_char

        # Setelah selesai mengiterasi melalui semua karakter, lalu
        # hasil dekripsi dengan nilai shift akan dicetak mulai dari 0 hingga 25.
        print("Berikut adalah hasil Descryption", shift, ":", decrypted_text)

# membuat menu untuk mempermudah user mencoba program
def caesar_cipher():
    print(" ___________________________")
    print("| Menu Program              |")
    print("| 0. Exit                   |")
    print("| 1. Encryption program     |")
    print("| 2. Decryption program     |")
    print(" ---------------------------")
    print("")
    option = input("Masukkan opsi: ")

    if option == "1":
        caesar_encryption()
    elif option == "2":
        caesar_decryption()
    elif option == "0":
        print("Terimakasih telah menggunakan program ini.")
        return
    else:
        print("Invalid! Coba Lagi!")
        caesar_cipher()

    caesar_cipher()

caesar_cipher()
```
